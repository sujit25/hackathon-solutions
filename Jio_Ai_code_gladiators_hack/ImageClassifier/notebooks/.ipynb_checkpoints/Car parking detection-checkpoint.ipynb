# Overview

This notebook is for training convolution neural network as image classifier for classifying patches of parking images as free/busy. If parking image is busy then it be blank else patch will be occupied by portion of car.

# Approach

Workflow for training image classification model.

- Read patch of images and their labels from filesystem. 
- Split image labels pairs into training and testing part where test size is 30% of all dataset.
- Feed images into batches using custom batch generator. Batch generator will read images for current batch, resize and reshape it & then feed it to cnn model.
- We have used keras with tensorflow backend for training cnn model.
- Similar to training batch generator, for evaluation purpose we feed test images using test batch generator and predict its labels.

## Summary
- We have trained convolutional neural network from sctrach on patches of images available from dataset.
- We have alculated different classification metrics for testing useful of our model. 
- We got accuracy of 95% with f1 score of 0.95.

```python
NUM_THREADS = 12
BATCH_SIZE = 64
IMG_HEIGHT = 150
IMG_WIDTH = 150
NUM_CHANNELS = 3
```

```python
def init_global_queue():
    queue = Queue()
    return queue
```

```python
def read_img(img_fpath):
    '''
    read an image from given img_fpath and reads into memory
    '''
    try:
        img =cv2.imread(img_fpath)
        #img = img.resize((IMG_HEIGHT,IMG_WIDTH), Image.ANTIALIAS)
        img = np.reshape(img, (-1, IMG_HEIGHT, IMG_WIDTH, NUM_CHANNELS))
        img = img/255
        return img
    except:
        return None
```

```python
def generate_batch(image_label_pairs):
    '''
    generate batch of image from images, labels pairs
    '''
    data_batch_indexes = list(np.random.choice(len(image_label_pairs), BATCH_SIZE))
    data_batch = [image_label_pairs[index] for index in data_batch_indexes]
    imgs_arrs = []
    imgs_labels = []
    #print("data batch size", len(data_batch))
    for img_fpath, img_label in data_batch:
        img_arr = read_img(img_fpath)
        if img_arr is not None:
            imgs_arrs.append(img_arr)
            imgs_labels.append(img_label)        
    concatenated_arr= np.concatenate(imgs_arrs, axis=1)
    concatenated_arr= np.reshape(concatenated_arr, (-1, IMG_HEIGHT, IMG_WIDTH, NUM_CHANNELS))
    #print(concatenated_arr.shape)
        #print(imgs_labels[:5])
    arr_len = min(len(concatenated_arr), len(img_labels))
    arr = concatenated_arr[:arr_len,:,:,:]
        #labels_arr = one_hot_encode_labels(np.array(img_labels).reshape(-1,1))
    labels_onehot_encoded = [np.array([0, 1]) if label == 1 else np.array([1, 0]) for label in imgs_labels]
    labels_arr = np.array(labels_onehot_encoded)
    #print(labels_arr.shape)
    labels = labels_arr[:arr_len]
    return arr, labels
```

```python
def batch_generator(image_label_pairs):    
    '''
    custom batch generator for generating batches of images & labels
    '''
    while True: 
        arr, labels = generate_batch(image_label_pairs)
        #print(arr.shape, labels.shape)                
        yield (arr, labels)        
```

### Training Image Classifier using Keras

```python
# CNN model
model = Sequential()

# -----------------------------------------------------------------------------------
# conv 1
model.add(Conv2D(16, (3,3), input_shape=(IMG_HEIGHT, IMG_WIDTH, NUM_CHANNELS)))       # input -N,150,150,3, output- N,148,148,16
#model.add(BatchNormalization(axis=3))
model.add(Activation('relu'))
#model.add(Dropout(0.5))

# max pool 1
model.add(MaxPooling2D(pool_size=(2,2),strides=2))                                   #input- N,148,148,16, output- N, 74,74,16

# -----------------------------------------------------------------------------------
# # conv 2
model.add(Conv2D(32, (3,3)))                                                         #input- N,74,74,16 output - N, 72,72,16
#model.add(BatchNormalization(axis=3))
model.add(Activation('relu'))
#model.add(Dropout(0.5))

# max pool 2
model.add(MaxPooling2D(pool_size=(2,2),strides=2))                                 #input - N,72,72,16, output- N,36,36,16
# -----------------------------------------------------------------------------------

# conv 3
model.add(Conv2D(48, (3,3)))                                                       #input - N,36,36,16, output- N,34,34,32
#model.add(BatchNormalization(axis=3))
model.add(Activation('relu'))
#model.add(Dropout(0.7))

# max pool 3
model.add(MaxPooling2D(pool_size=(2,2),strides=2))                                #input- N,34,34,32, output- N,17,17,32
# -----------------------------------------------------------------------------------

# # conv 4
model.add(Conv2D(64, (3,3)))                                                     #input- N,17,17,32, output- N,15,15,32
#model.add(BatchNormalization(axis=3))
model.add(Activation('relu'))
#model.add(Dropout(0.7))
# max pool 4
model.add(MaxPooling2D(pool_size=(2,2),strides=2))                              #input- N,15,15,32, output- N,7,7,32

# flatten
model.add(Flatten())                                                            # output- 1568

# fc layer 1
model.add(Dense(1024, activation='relu'))                                  

# fc layer 2
model.add(Dense(512, activation='relu'))

# fc layer 3
model.add(Dense(256, activation='relu'))

# fc layer 4
model.add(Dense(2, activation='softmax'))
```

```python
print(model.summary())
```

```python
# compile model for with softmax cross entropy and adam optimizer, set accuracy as parameter to evaluate
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# # train model on training data
#model_hist = model.fit_generator(, nb_epoch=10, verbose=1)
hist= model.fit_generator(batch_generator(train_img_label_pairs), steps_per_epoch=1000, epochs=5, verbose=1,max_queue_size=10, workers=8, use_multiprocessing=True, shuffle=True, initial_epoch=0)
```

### save model for future purpose

```python
def save_model(model):
    # serialize model to JSON
    model_json = model.to_json()
    with open("../trained_models/cnn_model/model.json", "w") as json_file:
        json_file.write(model_json)
    # serialize weights to HDF5
    model.save_weights("../trained_models/cnn_model/model.h5")
    print("Saved model to disk")
```

```python
save_model(model)
```
