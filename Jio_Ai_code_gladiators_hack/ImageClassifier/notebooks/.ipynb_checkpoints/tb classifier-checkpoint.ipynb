## Classification of Chest X-ray into normal/abnormal classes using deep learning

```python
import numpy as np
from os import listdir
from glob import glob
from PIL import Image
from skimage.transform import resize
import matplotlib.pyplot as plt
from os.path import basename,splitext
from os.path import exists

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn import metrics
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, BatchNormalization
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import np_utils
```

## Read dataset 

```python
dataset_montgomery_fpath = "/home/anton/Workspace/a_s/POCS/TB_Classification/NLM-MontgomeryCXRSet/MontgomerySet/CXR_png"
dataset_chinaset_fpath = "/home/anton/Workspace/a_s/POCS/TB_Classification/ChinaSet_AllFiles/CXR_png"
```

```python
montogmery_files = glob(dataset_montgomery_fpath + '/*.png')
chinaset_files = glob(dataset_chinaset_fpath + '/*.png')
```

```python
all_files = []
all_files.extend(montogmery_files)
all_files.extend(chinaset_files)
```

```python
print(len(all_files))
```

```python
IMG_SIZE= 512
NUM_CHANNELS=3
```

```python
def read_img(img_fpath):
    '''reads image from file path, converts into rgb and normalizes them'''
    img = Image.open(img_fpath)
    rgb_im = img.convert('RGB')
    rgb_img_arr = np.asarray(rgb_im)
    rgb_img_arr_resized = resize(rgb_img_arr, (IMG_SIZE, IMG_SIZE))
    rgb_img_arr_resized = ((rgb_img_arr_resized - rgb_img_arr_resized.mean())/rgb_img_arr_resized.std())
    return rgb_img_arr_resized
```

```python
def read_sample_img(img_fpath):
    img = read_img(img_fpath)
    label = splitext(basename(img_fpath))[0][-1]
    label_text = "normal" if label == '0' else 'abnormal'
    return img, label_text
```

```python
img0, img0_label = read_sample_img(all_files[0])
img10, img10_label = read_sample_img(all_files[9])
img100, img100_label = read_sample_img(all_files[99])
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,3,1)
plt.imshow(img0)
plt.title(img0_label)

plt.subplot(1,3,2)
plt.imshow(img10)
plt.title(img10_label)

plt.subplot(1,3,3)
plt.imshow(img100)
plt.title(img100_label)
```

## read image label pairs from all files 

```python
def generate_img_arrs_labels(all_fpaths):
    '''
    generate image & labels arrays
    '''
    if 
    img_arrs=[]
    labels =[]
    for img_fpath in all_fpaths:
        img_arr = read_img(img_fpath)
        label = splitext(basename(img_fpath))[0][-1]
        img_arrs.append(img_arr)
        labels.append(int(label))
    img_arr_stacked = np.stack(img_arrs)           
    img_labels = np.array(labels,dtype=np.int32)
    return img_arr_stacked, img_labels
```

```python
img_arr_stacked, img_labels = generate_img_arrs_labels(all_files)
```

```python
print(img_arr_stacked.shape)
print(img_labels.shape)
```

## one hot encode labels and save as np arrays

```python
def one_hot_encode_labels(img_labels_arr):
    ''' one hot encodes labels'''
    onehot_encoder = OneHotEncoder()
    return onehot_encoder.fit_transform(img_labels_arr.reshape(-1,1))
```

```python
np.save('img_arr_stacked.npy', img_arr_stacked)
np.save('img_labels.npy', img_labels)
```

```python
labels_ohe = one_hot_encode_labels(img_labels)
```

### Split Dataset into train and test set

```python
train_x, test_x, train_y, test_y = train_test_split(img_arr_stacked, labels_ohe, stratify= img_labels, test_size=0.1)
```

```python
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

## Build CNN model for Tuberculosis classification

```python
# CNN model
model = Sequential()

# -----------------------------------------------------------------------------------
# conv 1
model.add(Conv2D(16, (5,5), input_shape=(IMG_SIZE, IMG_SIZE, NUM_CHANNELS)))       # input -N,150,150,3, output- N,148,148,16
model.add(Activation('relu'))
#model.add(Dropout(0.5))

# max pool 1
model.add(MaxPooling2D(pool_size=(2,2),strides=2))                                   #input- N,148,148,16, output- N, 74,74,16

# -----------------------------------------------------------------------------------
# # conv 2
model.add(Conv2D(32, (3,3)))                                                         #input- N,74,74,16 output - N, 72,72,16
#model.add(BatchNormalization(axis=3))
model.add(Activation('relu'))
#model.add(Dropout(0.5))

# max pool 2
model.add(MaxPooling2D(pool_size=(2,2),strides=2))                                 #input - N,72,72,16, output- N,36,36,16
# -----------------------------------------------------------------------------------

# conv 3
model.add(Conv2D(64, (3,3)))                                                       #input - N,36,36,16, output- N,34,34,32
#model.add(BatchNormalization(axis=3))
model.add(Activation('relu'))
#model.add(Dropout(0.7))

# max pool 3
model.add(MaxPooling2D(pool_size=(2,2),strides=2))                                #input- N,34,34,32, output- N,17,17,32
# -----------------------------------------------------------------------------------

# # conv 4
model.add(Conv2D(128, (3,3)))                                                     #input- N,17,17,32, output- N,15,15,32
#model.add(BatchNormalization(axis=3))
model.add(Activation('relu'))
#model.add(Dropout(0.7))
# max pool 4
model.add(MaxPooling2D(pool_size=(2,2),strides=2))                              #input- N,15,15,32, output- N,7,7,32

# flatten
model.add(Flatten())                                                            # output- 1568

# fc layer 2
model.add(Dense(128, activation='relu'))

# fc layer 3
model.add(Dense(64, activation='relu'))

# fc layer 4
model.add(Dense(2, activation='softmax'))
```

```python
print(model.summary())
```

```python
# compile model for with softmax cross entropy and adam optimizer, set accuracy as parameter to evaluate
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
train_hist = model.fit(train_x, train_y, epochs=10, batch_size=32, verbose=1)
```

## visualization of training loss and accuracy

```python
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
plt.plot(range(len(train_hist.history['loss'])), train_hist.history['loss'])
plt.xlabel("epochs")
plt.ylabel("loss")
plt.title("variation of loss with epochs")

plt.subplot(1,2,2)
plt.plot(range(len(train_hist.history['acc'])), train_hist.history['acc'])
plt.xlabel("epochs")
plt.ylabel("accuracy")
plt.title("variation of accuracy with epochs")
```

```python
predictions = model.predict(test_x, verbose=1)
```

```python
def save_model(model):
    # serialize model to JSON
    model_json = model.to_json()
    with open("../model/model.json", "w") as json_file:
        json_file.write(model_json)
    # serialize weights to HDF5
    model.save_weights("../model/model.h5")
    print("Saved model to disk")
```

```python
save_model(model)
```

## Computation of Metrics

```python
predicted_labels = np.argmax(predictions, axis=1)
actual_test_labels = np.argmax(test_y, axis=1)
```

```python
print("accuracy", metrics.accuracy_score(actual_test_labels, predicted_labels))
print("f1 score", metrics.f1_score(actual_test_labels, predicted_labels))
print("precison score:", metrics.precision_score(actual_test_labels, predicted_labels))
print("recall score:", metrics.recall_score(actual_test_labels, predicted_labels))
```
